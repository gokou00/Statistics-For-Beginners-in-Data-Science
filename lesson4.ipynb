# Pandas, Matplotlib, and Seaborn for Statistical Visualization

# Pie Charts

```python
import matplotlib.pyplot as plt
labels = ["Green","Red","Black","Yellow"]
```

```python
#Assigning a size and color for each part of the pie chart

sizes = [38.4,40.6,20.7,10.3]
colors = ["green","red","black","yellow"]
```

```python
#Calling the pie function from Matplotlib. We pass in the sizes and the colors. Finally we insert a legend along with the axis 

patches,texts = plt.pie(sizes,colors=colors)
plt.legend(patches,labels,loc="best")
plt.axis("equal")
plt.tight_layout()
plt.show()
```

```python
# Plotting the same chart but using the pandas library

import pandas as pd

df = pd.DataFrame([38.4,40.6,20.7,10.3],index=["blue","orange","green","red"],columns=["pie chart"])

#making the plot

df.plot(kind="pie",subplots=True,figsize=(8,8))
```

# Bar Graphs

```python
# Plotting the same data using both seaborn and Matplotlib

#starting with seaborn

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Setting up Matplotlib figure and generate some dummy data

f,ax1 = plt.subplots(1,1)
x = np.array(list(range(1,11)))
y1 = np.arange(1,11)

#plotting using seaborn
sns.barplot(x=x,y=y1,palette="Blues",ax=ax1)
plt.show()
```

```python
#Plotting using Matplotlib

plt.bar(x,y1,label="Bar1")
```

# Box Plots

```python
# Generating random data

s = np.random.rand(50) * 100

c = np.ones(25) * 50

high = np.random.rand(10) * 100 + 100
low = np.random.rand(10) * -100
data = np.concatenate((s,c,high,low))
df = pd.DataFrame(data)
```

```python
# Printing the five-number summary

df.describe()
```

```python
# Plotting the data as a box plot using Matplotlib

fig1,ax1 = plt.subplots()
ax1.set_title("Box Plot")
ax1.boxplot(data)
```

```python
# plotting the same data using seaborn 

ax = sns.boxplot(x=df,orient="v")
```

# Frequency Table and Histogram

```python
df = pd.DataFrame({"names":["Ahmed","SciencesAI","Ahmed","Ahmed","Jack","SciencesAI","SciencesAI","Rose","Jack","Ahmed"]})

#Creating the frequency table using the value_counts functions Pandas.

count = df["names"].value_counts()
print(count)
```

```python
# Visualizing it as a histogram

sns.catplot(x="names",kind="count",data=df)
```

# Time-Series Charts

```python
from dateutil.parser import parse
plt.rcParams.update({"figure.figsize":(10,7),"figure.dpi":120})

#importing the dataset as a dataframe from over the internet

df = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/a10.csv",parse_dates=["date"])
df.head()
```

```python
# Now plotting the data as a time series

plt.figure(figsize=(16,5),dpi=100)
plt.plot(df["date"],df["value"])
plt.gca().set(xlabel="Date",ylabel="Sales")
plt.show()
```
