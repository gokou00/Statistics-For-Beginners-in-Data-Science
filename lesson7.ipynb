# Confidence Interval for a Population Mean

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Generating 100000 data points with a mean of 4 and a standard deviation of 2.82
shape,scale = 2.0,2.0
s = np.random.gamma(shape,scale,100000)
mu = shape*scale
sigma = scale*np.sqrt(shape)
print(mu)
print(sigma)
```

```python
# Creating a mean sample array with a sample size of 500

sample_mean = []
sample_size = 200
for j in range(5000):
    random_sample = random.choices(s,k=sample_size)
    sample_mean.append(sum(random_sample)/len(random_sample))
```

```python
# Now plotting the data

plt.figure(figsize=(20,10))
plt.hist(sample_mean,200,density=True,color="lightgreen")
plt.show()
```

```python
#Finding the confidence interval on the graph

plt.figure(figsize=(20,10))
plt.hist(sample_mean,200,density=True,color="lightgreen")
plt.plot([mu,mu],[0,3.2],"k-",lw=4,color="blue")
plt.plot([mu-(1.96*sigma/np.sqrt(sample_size)),mu-(1.96*sigma/np.sqrt(sample_size))],[0,3.2],"k-",lw=2,color="black")
plt.plot([mu+(1.96*sigma/np.sqrt(sample_size)),mu+(1.96*sigma/np.sqrt(sample_size))],[0,3.2],"k-",lw=2,color="black")
plt.show()
```

# Confidence Interval for a Population Proportion

```python
import statsmodels.stats.proportion as sm
```

```python
z_value = 1.96
population_proportion = 0.9
number_of_observations = 600
```

```python
# Calculate the standard error 

standard_error = np.sqrt((population_proportion*(1-population_proportion))/number_of_observations)
print(standard_error)
```

```python
# Calculate the lower and upper bound using the above equation

lower_bound = population_proportion - (standard_error*z_value)
upper_bound = population_proportion + (standard_error*z_value)
lower_bound,upper_bound
```

```python
#Calculating the same thing using StatsModels

sm.proportion_confint(number_of_observations*population_proportion,number_of_observations)
```
